```python
from data import eda                         # Importa o módulo personalizado 'eda' do pacote 'data' que contém a classe Eda
from IPython.display import display_markdown # Importa função para exibir markdown formatado em notebooks Jupyter

path = 'dataset/pokemon.csv' # Define o caminho para o arquivo CSV do dataset de Pokémon
df = eda.Eda(path)           # Cria uma instância da classe Eda passando o caminho do arquivo como parâmetro
load = df.load_dataset()     # Chama o method para carregar o dataset e retorna status da operação

if load: # Se houve erro no carregamento (load=True), exibe mensagem e encerra programa
    print("Programa finalizado sem êxito.")
    exit()
else:
    title = 'ANÁLISE EXPLORATÓRIA DE DADOS - POKEMON WITH STATS' # Define o título principal da análise
    display_markdown(f"# **{title}**", raw=True)                 # Exibe o título formatado como cabeçalho markdown no notebook

    # df.report(df.df_L0, title.title()+' (Level 0)')            # Gera relatório completo dos dados originais (Level 0)
    df.statistics(df.df_L0, '1', '(Level 0)')                    # Gera estatísticas descritivas dos dados originais (Level 0)
    df.set_target('Type 2')                                      # Define a coluna 'Type 2' como variável target/dependente para análise
    # df.set_target('Total')                                       # Define a coluna 'Total' como variável target/dependente para análise
    validation = df.analysis_target()                            # Analisa a variável target e retorna status de validação

    if validation: # Se houve erro na análise do target, encerra o programa
        print("Programa finalizado sem êxito.")
        exit()
    else:
        df.transformation()                                      # Aplica transformações e limpeza nos dados
        # df.report(df.df_L1, title.title()+' (Level 1)')        # Gera relatório dos dados após transformação
        df.statistics(df.df_L1, '3', '(Level 1)')                # Gera estatísticas dos dados transformados (Level 1)
        df.summary(df.df_L1)                                     # Exibe resumo detalhado do dataset transformado
        df.preview_target(df.df_L1)                              # Visualiza a distribuição e características da variável target
        df.analysis_columns(df.df_L1)                            # Analisa cada coluna individualmente com gráficos e estatísticas
        df.correlation_heatmap(df.df_L1)                         # Gera heatmap de correlação entre as variáveis numéricas
```
